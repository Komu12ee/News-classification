```python
import pandas as pd

# Replace 'cricket_news.xlsx' with your file path
df = pd.read_excel('/content/politics18.csv.xlsx')

# Show the first 5 rows to check the data
print(df.head())
```

```python
df = df.reset_index(drop=True)
```

```python
print(df.columns)
# Drop all columns whose name contains 'Unnamed'
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

print(df.columns)  # Verify
```

```python
# Remove duplicate rows based on 'Title' and 'Content'
df = df.drop_duplicates(subset=['title', 'content'])

print(f"After removing duplicates, total rows: {len(df)}")
```

```python
# Check for missing values
print(df.isnull().sum())

# Drop rows where 'headline' or 'Content' is missing
df = df.dropna(subset=['title', 'content'])

print(f"After dropping missing values, total rows: {len(df)}")
```

```python
# Function to clean text
import re

def clean_text(text):
    text = str(text)  # Ensure it's a string
    text = text.strip()  # Remove leading/trailing spaces
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with single space
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation/special characters
    return text

# Apply cleaning to 'Title' and 'Content'
df['headline'] = df['title'].apply(clean_text)
df['content'] = df['content'].apply(clean_text)

print(df.head())
```

```python
df.head()
```

```python
# Convert all text to lowercase
df['headline'] = df['headline'].str.lower()
df['content'] = df['content'].str.lower()
```

```python
df.head()
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return ' '.join(words)

df['headline'] = df['headline'].apply(remove_stopwords)
df['content'] = df['content'].apply(remove_stopwords)
```

```python
df.head()
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    words = text.split()
    words = [lemmatizer.lemmatize(w) for w in words]
    return ' '.join(words)

df['content'] = df['content'].apply(lemmatize_text)
```

```python
df.head()
```

```python
# Combine headline and content into one column for modeling
df['text'] = df['headline'] + ' ' + df['content']

# View the first rows
print(df[['text', 'category']].head())
```

```python
df = df.drop(columns=['headline', 'content','title'])
```

```python
import re

def clean_special_chars(text):
    # Replace unusual characters with a space
    text = re.sub(r'[^A-Za-z0-9\s.,!?]', ' ', str(text))
    # Replace multiple spaces with single space
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

# Apply to your 'text' column
df['text'] = df['text'].apply(clean_special_chars)
```

```python
df.head()
```

```python
df['text'] = df['text'].str.lower()
```

```python
df.to_csv('politic_news_cleaned.csv', index=False, header=True)
```
